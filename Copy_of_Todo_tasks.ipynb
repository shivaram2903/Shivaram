```python
todo_tasks = {

}

def menu_option():
    print("\nMenu\n")
    print("Enter 1 to add new task.\n")
    print("Enter 2 to Delete a task. \n")
    print("Enter 3 to display the task. \n")
    print("Enter 4 to mark as complete a task. \n")
    print("Enter 5 to Exit\n")

    option = int(input("Enter your choice : "))

    print("\n")
    result = choose_option(option)





def add_task():
    task_name = input("add required task : ")
    res = ""
    for char in task_name:
        res += char.lower()
    todo_tasks[res] = False

    print("\nThe task * {} * been added successfuly.".format(res))

    menu_option()

def delete_task():
    deleted_task_name = input("Please Enter the Task name to be removed: ")
    res = ""
    for char in deleted_task_name:
        res += char.lower()
    tasks_list = todo_tasks.keys()
    tasks_list = list(tasks_list)

    if res not in tasks_list:
        print("\nThe Task name {} is not in the Todo List!\n".format(res))
    else:
        del todo_tasks[res]
        print("\nThe task {} has been deleted successfully!\n".format(res))

    menu_option()









def view_task():
    tasks_list = todo_tasks.keys()
    tasks_list = list(tasks_list)

    length_of_tasks_list = len(tasks_list)

    if length_of_tasks_list == 0:
        print("The Todo List is empty!\n")
    else:
        print("The Todo List consists of {} Todo Items.\nThey are as follows:\n".format(length_of_tasks_list))
        serial_num = 0
        for i in tasks_list:
            status_of_task = ""
            serial_num = serial_num + 1

            if todo_tasks[i] == False:
                status_of_task = "Not Completed!"
            else:
                status_of_task = "Completed!"

            print("{}. {} ---> status of the task: {}\n".format(serial_num,i,status_of_task))
    menu_option()




def mark_as_complete():
    task_name = input("Enter the task name to change its status: ")
    task_name = task_name.lower()
    change_status_of_task = input("\nHave you completed the task? (YES/NO): ")
    change_status_of_task = change_status_of_task.lower()
    if change_status_of_task == "no" or change_status_of_task == "n":
        todo_tasks[task_name] = False
    elif change_status_of_task == "yes" or change_status_of_task == "y":
        todo_tasks[task_name] = True

    print("\nThe status of the task has been changed!\n")

    menu_option()


def exit_function():
    try:
        exit()

    finally:
        print("\nYou have Exited from the Todo List Tasks successfully!\n")








def choose_option(option):
    value = option

    if(value == 1):
        New_task = add_task()

    elif(value == 2):
        remove_task = delete_task()

    elif(value == 3):
        display_task = view_task()

    elif(value == 4):
        complete_task_status = mark_as_complete()

    elif(value == 5):
        exit_from_list = exit_function()

    else:
        print("\n")
        print("You entered the wrong option !! \n")
        print("Please select from the below Choices.  \n")

        menu_option()

menu_option()
```
